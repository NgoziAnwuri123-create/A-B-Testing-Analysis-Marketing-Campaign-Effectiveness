## 1.1 What is NumPy and Why Do We Need It?

NumPy (Numerical Python) is the **core library for numerical computing** in Python. Before NumPy, doing math on large lists of numbers was slow and awkward. NumPy introduces the **ndarray** — a super-powered array that:

- Stores numbers in a compact, memory-efficient block (unlike Python lists which store references)
- Performs mathematical operations on **all elements at once** (called *vectorisation*)
- Is 10–100× faster than equivalent Python list operations
- Forms the foundation of pandas, scikit-learn, TensorFlow, and almost every data tool

**Real-world analogy:** Think of a Python list like a bag of shopping — each item wrapped separately. NumPy arrays are like a tray of eggs — same shape, same type, all in one place, easy to count and process together.

---


## 1.2 Importing NumPy

We import NumPy with the alias `np` — this is a universal convention. You will see `np.` everywhere in data science code.


```python
import numpy as np
print('NumPy version:', np.__version__)
```

## 1.3 Creating NumPy Arrays

There are several ways to create arrays. The most direct is `np.array()` from a Python list.

**Key mental model:** A NumPy array has:
- **shape** — the dimensions (rows × columns etc.)
- **dtype** — the data type of every element (int, float, etc.)
- **ndim** — the number of dimensions

We'll start with 1D (a single row of numbers), then move to 2D (a table of numbers like a spreadsheet).


```python
# 1D array — like a single column of values
temperatures = np.array([36.6, 37.1, 36.9, 38.2, 37.0])
print("Array:", temperatures)
print("Shape:", temperatures.shape)
print("Data type:", temperatures.dtype)
print("Dimensions:", temperatures.ndim)
```

```python
# 2D array — like a mini spreadsheet (rows × columns)
# Each row = one student's test scores [Math, English, Science]
scores = np.array([
    [85, 90, 78],
    [92, 88, 95],
    [70, 75, 80],
    [88, 82, 91]
])
print("Scores array:\n", scores)
print("Shape:", scores.shape, "← 4 students, 3 subjects")
```

### Useful Array Creation Shortcuts

You don't always type values manually. NumPy has helpers:


```python
# Evenly spaced values
hours = np.arange(0, 24, 1)          # 0 to 23, step 1
print("Hours in a day:", hours)

# 5 evenly spaced points between 0 and 1
percentages = np.linspace(0, 1, 5)
print("Evenly spaced:", percentages)

# Arrays of zeros and ones (useful for initialising)
zeros = np.zeros((3, 4))
print("Zeros grid (3x4):\n", zeros)
```

## 1.4 Indexing and Slicing

Getting specific values out of an array uses the same square-bracket notation as Python lists — but NumPy extends it for multiple dimensions.

**Rule of thumb:** `array[row, column]` for 2D. Use `:` to mean "all of this dimension".


```python
# 1D indexing — same as Python lists
temps = np.array([36.6, 37.1, 36.9, 38.2, 37.0, 36.8])
print("First reading:", temps[0])
print("Last reading:", temps[-1])
print("First three readings:", temps[:3])

# Boolean indexing — VERY powerful: "give me elements that match a condition"
fever_threshold = 37.5
fevers = temps[temps > fever_threshold]
print("Fever readings (>37.5°C):", fevers)
```

```python
# 2D indexing on our student scores
scores = np.array([
    [85, 90, 78],   # Student 0
    [92, 88, 95],   # Student 1
    [70, 75, 80],   # Student 2
    [88, 82, 91]    # Student 3
])
# Column labels: [Math, English, Science]

print("Student 1's scores:", scores[1])           # entire row
print("All Math scores:", scores[:, 0])            # entire first column
print("Top-left 2x2 block:\n", scores[:2, :2])    # slicing both dims
```

## 1.5 Vectorised Operations — The Power of NumPy

This is where NumPy shines. Instead of writing a loop, you apply an operation to the **entire array at once**.

**Why it matters:** Internally, NumPy executes this in compiled C code. No Python interpreter overhead. For a million numbers, this is the difference between 0.001 seconds and 1 second.


```python
import time

# Compare: Python list loop vs NumPy vectorised operation
big_list = list(range(1_000_000))
big_array = np.array(big_list)

# Python loop
start = time.time()
squared_list = [x**2 for x in big_list]
python_time = time.time() - start

# NumPy vectorised
start = time.time()
squared_array = big_array ** 2
numpy_time = time.time() - start

print(f"Python loop: {python_time:.4f} seconds")
print(f"NumPy vectorised: {numpy_time:.4f} seconds")
print(f"NumPy is ~{python_time/numpy_time:.0f}x faster!")
```

```python
# Arithmetic on arrays — element-wise by default
prices = np.array([100, 250, 89, 430, 175])   # product prices in USD
discount = 0.10                                # 10% off

discounted = prices * (1 - discount)
print("Original prices:  ", prices)
print("After 10% discount:", discounted)

# Add VAT (7.5%) to discounted prices
with_vat = discounted * 1.075
print("After adding 7.5% VAT:", np.round(with_vat, 2))
```

## 1.6 Aggregation Functions

NumPy has built-in statistical functions that work on whole arrays — or along specific axes of 2D arrays.

- `axis=0` → operate **down the rows** (collapse rows, keep columns)
- `axis=1` → operate **across columns** (collapse columns, keep rows)


```python
# Student scores again: rows=students, cols=[Math, English, Science]
scores = np.array([
    [85, 90, 78],
    [92, 88, 95],
    [70, 75, 80],
    [88, 82, 91]
])

print("Mean score per subject (axis=0):", np.mean(scores, axis=0))
print("  → [Math avg, English avg, Science avg]")
print()
print("Total score per student (axis=1):", np.sum(scores, axis=1))
print("  → [Student0 total, Student1 total, ...]")
print()
print("Highest mark in the whole class:", np.max(scores))
print("Which student got highest total?", np.argmax(np.sum(scores, axis=1)))
```

## 1.7 Reshaping Arrays

Sometimes data comes in one shape but you need it in another. `reshape()` rearranges the elements without copying data — the total number of elements must stay the same.

A value of `-1` in reshape means "figure out this dimension for me".


```python
# 12 monthly sales figures → reshape into 4 quarters × 3 months
monthly_sales = np.array([
    12000, 15000, 9500,   # Q1
    18000, 21000, 17500,  # Q2
    14000, 16500, 19000,  # Q3
    22000, 25000, 23500   # Q4
])

quarterly = monthly_sales.reshape(4, 3)
print("Monthly sales reshaped to quarterly view:")
print(quarterly)
print()
print("Quarter totals:", np.sum(quarterly, axis=1))
print("Best quarter: Q" + str(np.argmax(np.sum(quarterly, axis=1)) + 1))
```

## 🏥 Real-World Task Hospital Patient Vitals Analysis

**Scenario:** You are a data analyst at a Lagos hospital. The system has exported one week of patient vitals for 8 patients. Your job is to flag anomalies, summarise trends, and produce the numbers a doctor needs at a glance.

**Data:** Each row is a patient. Columns are: `[Monday, Tuesday, Wednesday, Thursday, Friday, Saturday, Sunday]` body temperatures in °C.

**Your tasks:**
1. Find each patient's average temperature over the week
2. Identify which patients had a fever (>37.5°C) on any day
3. Find the hottest recorded temperature each day across all patients
4. Normalise the data: express each reading as a deviation from 37.0°C (normal body temp)

---


```python
# ── Setup ────────────────────────────────────────────────────────────
np.random.seed(42)  # so everyone gets the same "random" data

# 8 patients × 7 days of temperature readings
patient_ids = ["P001","P002","P003","P004","P005","P006","P007","P008"]
days = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]

vitals = np.array([
    [36.8, 37.0, 37.2, 38.5, 38.9, 37.8, 37.1],  # P001
    [36.5, 36.6, 36.7, 36.8, 36.9, 36.5, 36.6],  # P002
    [37.1, 37.3, 38.0, 39.1, 39.4, 38.7, 37.9],  # P003
    [36.9, 36.9, 37.0, 37.1, 37.0, 36.8, 36.9],  # P004
    [37.0, 37.8, 38.5, 38.8, 37.6, 37.2, 37.0],  # P005
    [36.4, 36.5, 36.5, 36.6, 36.4, 36.3, 36.4],  # P006
    [38.1, 38.4, 39.0, 39.5, 38.8, 38.2, 37.8],  # P007
    [37.0, 37.0, 37.1, 37.0, 37.2, 37.1, 37.0],  # P008
])

print("Patient vitals (°C) — rows=patients, cols=days")
print(vitals)
```

```python
# ── Task 1: Average temperature per patient ──────────────────────────
avg_temps = np.mean(vitals, axis=1)   # axis=1 → average across days (per patient)
print("Average temperature per patient:")
for pid, avg in zip(patient_ids, avg_temps):
    print(f"  {pid}: {avg:.2f}°C")
```

```python
# ── Task 2: Flag patients who had fever (>37.5°C) on any day ─────────
FEVER = 37.5
had_fever = np.any(vitals > FEVER, axis=1)   # True/False per patient
print("Fever alert (True = had fever at least once):")
for pid, flag in zip(patient_ids, had_fever):
    status = " FEVER" if flag else " OK"
    print(f"  {pid}: {status}")
```

```python
# ── Task 3: Peak temperature each day across all patients ────────────
daily_peak = np.max(vitals, axis=0)   # axis=0 → max down patients (per day)
print("Daily peak temperature across all patients:")
for day, peak in zip(days, daily_peak):
    print(f"  {day}: {peak:.1f}°C")
    
worst_day = days[np.argmax(daily_peak)]
print(f"\nHighest readings were on: {worst_day}")
```

```python
# ── Task 4: Deviation from normal (37.0°C) ───────────────────────────
NORMAL = 37.0
deviation = vitals - NORMAL    # vectorised subtraction — no loop needed!

print("Temperature deviation from 37.0°C (+ means above normal):")
print(np.round(deviation, 1))
print()
# Find the most extreme deviation (absolute value)
max_dev_idx = np.unravel_index(np.argmax(np.abs(deviation)), deviation.shape)
print(f"Most extreme deviation: Patient {patient_ids[max_dev_idx[0]]} on {days[max_dev_idx[1]]}")
print(f"  Actual temp: {vitals[max_dev_idx]:.1f}°C  |  Deviation: {deviation[max_dev_idx]:+.1f}°C")
```

## 2.1 What is Pandas?

Pandas is built **on top of NumPy** and adds the ability to work with **labelled, heterogeneous tabular data** — think spreadsheets and SQL tables, but inside Python.

Key differences from NumPy:

| NumPy | Pandas |
|---|---|
| All values same type | Mixed types (strings, numbers, dates) |
| Accessed by position | Accessed by **name** (column/index labels) |
| Best for pure maths | Best for real-world data analysis |
| Rows and columns unnamed | Named rows (index) and columns |

**Two core structures:**
- **Series** — a single labelled column (1D)
- **DataFrame** — a table of labelled columns (2D) ← you'll use this 90% of the time


## 2.2 Importing Pandas


```python
import pandas as pd
import numpy as np
print('Pandas version:', pd.__version__)
```

## 2.3 The Series — A Labelled Column

A Series is a 1D array with an **index** (labels for each value). If you don't provide an index, pandas uses 0, 1, 2...


```python
# Creating a Series with explicit labels
city_populations = pd.Series(
    [15_900_000, 4_500_000, 2_300_000, 1_200_000, 900_000],
    index=["Lagos", "Kano", "Ibadan", "Abuja", "Port Harcourt"],
    name="Population"
)
print(city_populations)
print()
print("Lagos population:", city_populations["Lagos"])
print("Cities with pop > 2M:\n", city_populations[city_populations > 2_000_000])
```

## 2.4 The DataFrame — Your Digital Spreadsheet

A DataFrame is a 2D table. Think of it as a dictionary of Series all sharing the same index.

**How to create one:**
- From a dictionary of lists (most common)
- From a CSV file (`pd.read_csv()`) ← how you'll use it in real projects
- From a NumPy array


```python
# Create a DataFrame from a dictionary
employees = pd.DataFrame({
    "name":       ["Amara", "Chidi", "Fatima", "Emeka", "Ngozi"],
    "department": ["Sales", "Engineering", "HR", "Engineering", "Sales"],
    "salary":     [450_000, 720_000, 380_000, 850_000, 490_000],
    "years":      [3, 5, 2, 7, 4],
    "remote":     [True, True, False, True, False]
})

print(employees)
print()
print("Shape:", employees.shape, "← (rows, columns)")
print("Columns:", list(employees.columns))
```

## 2.5 Exploring a DataFrame — Your First 5 Commands

Whenever you receive a new dataset, these five commands should be your reflex:


```python
# The 5 must-know exploration commands
print("─── .head() — first 5 rows ───")
print(employees.head())
print()
print("─── .info() — column types & nulls ───")
employees.info()
```

```python
print("─── .describe() — statistics for numeric columns ───")
print(employees.describe())
```

```python
print("─── .dtypes — data type of each column ───")
print(employees.dtypes)
print()
print("─── .value_counts() — frequency of categories ───")
print(employees["department"].value_counts())
```

## 2.6 Selecting Columns and Rows

Pandas gives you two ways to select data:

- **`df["column"]`** → select a single column (returns a Series)
- **`df[["col1","col2"]]`** → select multiple columns (returns a DataFrame)
- **`.loc[]`** → select by **label** (name)
- **`.iloc[]`** → select by **integer position** (like NumPy)


```python
# Selecting columns
print("Salary column (Series):")
print(employees["salary"])
print()
print("Name + Salary columns (DataFrame):")
print(employees[["name","salary"]])
```

```python
# Selecting rows with .loc (by label/index)
print("Row at index 2:", employees.loc[2].to_dict())
print()

# .iloc (by position, just like NumPy)
print("First 3 rows, columns 0-2:")
print(employees.iloc[:3, :3])
```

## 2.7 Filtering Rows — Boolean Indexing

This is one of the most important Pandas skills. You create a **boolean mask** (a column of True/False) and use it to filter rows.

You can chain conditions with:
- `&` for AND
- `|` for OR
- `~` for NOT

> ⚠️ Always wrap each condition in parentheses when combining!


```python
# Single condition: Engineering staff only
engineers = employees[employees["department"] == "Engineering"]
print("Engineering staff:")
print(engineers)
```

```python
# Multiple conditions: remote workers earning more than 600k
high_earners_remote = employees[
    (employees["remote"] == True) &
    (employees["salary"] > 600_000)
]
print("Remote workers earning > ₦600k:")
print(high_earners_remote)
```

## 2.8 Creating New Columns and Transforming Data

You can add a new column by assigning to `df["new_column"]`. NumPy-style vectorised operations work directly on columns.

`.apply()` lets you apply any Python function to a column row-by-row — use it when vectorised math isn't enough.


```python
# Add calculated columns
employees["monthly_salary"] = employees["salary"] / 12
employees["bonus_10pct"] = employees["salary"] * 0.10

print(employees[["name","salary","monthly_salary","bonus_10pct"]])
```

```python
# .apply() — run any function across a column
def experience_band(years):
    if years <= 2:
        return "Junior"
    elif years <= 5:
        return "Mid-level"
    else:
        return "Senior"

employees["level"] = employees["years"].apply(experience_band)
print(employees[["name","years","level"]])
```

## 2.9 GroupBy — Aggregating by Category

`groupby()` is the pandas equivalent of SQL's `GROUP BY`. It splits the data by a category, applies a function to each group, and combines the results.

**The pattern:** `df.groupby("column")["value_column"].aggregation()`


```python
# Average salary by department
print("Average salary per department:")
print(employees.groupby("department")["salary"].mean().round(0))
print()

# Multiple aggregations at once
print("Department summary:")
dept_summary = employees.groupby("department").agg(
    headcount=("name", "count"),
    avg_salary=("salary", "mean"),
    avg_years=("years", "mean")
).round(1)
print(dept_summary)
```

## 2.10 Handling Missing Data

Real-world data is messy. Pandas represents missing values as `NaN` (Not a Number). Key operations:

- `.isnull()` / `.notnull()` — detect missing values
- `.fillna(value)` — fill NaN with a default
- `.dropna()` — drop rows (or columns) with any NaN


```python
# Introduce some missing data (common in real datasets)
df_messy = pd.DataFrame({
    "product": ["Phone", "Laptop", "Tablet", "Watch", "Earbuds"],
    "price":   [150_000, 450_000, None, 85_000, None],
    "rating":  [4.5, 4.8, 4.2, None, 4.6],
    "stock":   [50, 12, 0, None, 200]
})
print("Dataset with missing values:")
print(df_messy)
print()
print("Missing values per column:")
print(df_messy.isnull().sum())
```

```python
# Strategy 1: Fill NaN with column mean (for price)
df_messy["price"] = df_messy["price"].fillna(df_messy["price"].mean())

# Strategy 2: Fill NaN with a specific default (0 for stock)
df_messy["stock"] = df_messy["stock"].fillna(0)

# Strategy 3: Drop rows where rating is missing
df_clean = df_messy.dropna(subset=["rating"])

print("After cleaning:")
print(df_clean)
print()
print("No more missing values:", df_clean.isnull().sum().sum() == 0)
```

## 2.11 Sorting and Ranking

`.sort_values()` sorts by one or more columns. `ascending=False` for descending order.


```python
# Sort employees by salary, highest first
top_earners = employees.sort_values("salary", ascending=False)
print("Employees ranked by salary:")
print(top_earners[["name","department","salary","level"]].reset_index(drop=True))
```

```python
# Sort by multiple columns: department first, then salary
multi_sort = employees.sort_values(
    ["department","salary"], 
    ascending=[True, False]
)
print("Sorted by department, then salary (desc):")
print(multi_sort[["name","department","salary"]].reset_index(drop=True))
```

## 🛒 Real-World Task — E-Commerce Sales Analysis

**Scenario:** You work as a data analyst for an online store selling electronics across Nigeria. You have been handed a month's transaction data and asked to produce a report for the business team before Friday's meeting.

**Questions the business wants answered:**
1. What are the total sales and average order value per product category?
2. Which city generates the most revenue?
3. What is the day-over-day sales trend for the month?


```python
# ── Build the dataset ─────────────────────────────────────────────────
np.random.seed(99)
n = 200

categories = ["Phones", "Laptops", "Accessories", "Tablets", "Audio"]
cities = ["Lagos", "Abuja", "Kano", "Port Harcourt", "Ibadan"]

orders = pd.DataFrame({
    "order_id":   [f"ORD-{1000+i}" for i in range(n)],
    "date":       pd.date_range("2024-01-01", periods=n, freq="4h").normalize(),
    "city":       np.random.choice(cities, n, p=[0.40, 0.20, 0.15, 0.15, 0.10]),
    "category":   np.random.choice(categories, n, p=[0.35, 0.25, 0.20, 0.12, 0.08]),
    "quantity":   np.random.randint(1, 15, n),
    "unit_price": np.random.choice([15000,45000,89000,125000,220000,350000], n),
    "discount_pct": np.round(np.random.uniform(0, 0.40, n), 2),
})

orders["revenue"] = orders["quantity"] * orders["unit_price"] * (1 - orders["discount_pct"])
orders["revenue"] = orders["revenue"].round(2)

print("Sample of orders data:")
print(orders.head(8))
print(f"\nTotal records: {len(orders)}")
```

```python
# ── Task 1: Sales summary per category ───────────────────────────────
category_summary = orders.groupby("category").agg(
    total_orders   = ("order_id", "count"),
    total_revenue  = ("revenue", "sum"),
    avg_order_value= ("revenue", "mean"),
    avg_discount   = ("discount_pct", "mean")
).sort_values("total_revenue", ascending=False).round(2)

print("Sales Summary by Category")
print("="*60)
print(category_summary.to_string())
```

```python
# ── Task 2: Revenue by city ───────────────────────────────────────────
city_revenue = (orders
    .groupby("city")["revenue"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
city_revenue.columns = ["City", "Total Revenue (₦)"]
city_revenue["Share %"] = (city_revenue["Total Revenue (₦)"] / 
                            city_revenue["Total Revenue (₦)"].sum() * 100).round(1)

print("Revenue by City")
print("="*40)
print(city_revenue.to_string(index=False))
print(f"\nTop city: {city_revenue.iloc[0]['City']} ({city_revenue.iloc[0]['Share %']}% of revenue)")
```

```python
# ── Task 3: Day-over-day sales trend ─────────────────────────────────
daily = (orders
    .groupby("date")["revenue"]
    .sum()
    .reset_index()
    .rename(columns={"revenue":"daily_revenue"})
)
daily["7day_avg"] = daily["daily_revenue"].rolling(window=7, min_periods=1).mean().round(0)
daily["day_name"] = daily["date"].dt.day_name()

print("Daily Revenue (first 14 days shown)")
print("="*55)
print(daily.head(14)[["date","day_name","daily_revenue","7day_avg"]].to_string(index=False))
print(f"\nAverage daily revenue: ₦{daily['daily_revenue'].mean():,.0f}")
print(f"Best day:  {daily.loc[daily['daily_revenue'].idxmax(), 'date'].date()} — ₦{daily['daily_revenue'].max():,.0f}")
print(f"Worst day: {daily.loc[daily['daily_revenue'].idxmin(), 'date'].date()} — ₦{daily['daily_revenue'].min():,.0f}")
```
